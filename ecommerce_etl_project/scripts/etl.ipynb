# Extract, Transform, Load (ETL)

ETL is the process of moving data from source systems into a target system (e.g., a data warehouse) by extracting raw data, transforming it into a consistent, analyzable format, and loading it into the destination.

- Extract: ingest data from databases, APIs, files, or streams.
- Transform: clean, validate, normalize, enrich, and aggregate data; apply business rules.
- Load: write transformed data to the target system, ensuring idempotency and integrity.

Best practices: handle errors and schema changes, automate and schedule pipelines, monitor and log processing.
